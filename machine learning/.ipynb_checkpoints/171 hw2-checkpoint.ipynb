```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import keras
import matplotlib.pyplot as plt 
#from keras.callbacks import LambdaCallback
```

```python
df = pd.read_table("yeast.data", sep="\s+", header=None)
```

```python
df.columns = ['Sequence_Name', 'mcg', 'gvh', 'alm', 'mit', 'erl', 'pox', 'vac', 'nuc', 'site']

# isolate ERL class since it will be eliminated as outliers
df_ERL = df[df['site']=='ERL']
df.drop(df_ERL.index,inplace=True)
df = df.reset_index(drop=True)
```

```python
fea = df[['mcg', 'gvh', 'alm', 'mit', 'erl', 'pox', 'vac', 'nuc']].values
```

## Isolation forest outlier detector

```python
# isolation forest outlier detector
clf = IsolationForest(contamination='auto',behaviour="new")
arr = clf.fit_predict(fea)
```

```python
itemindex = np.where(arr==-1)
outlierindex = itemindex[0]
outlierindex
```

## Local outlier factor

```python
from sklearn.neighbors import LocalOutlierFactor
neigh = LocalOutlierFactor(contamination='auto')
outlierindex2 = neigh.fit_predict(fea)
outlierindex2 = np.where(outlierindex2==-1)[0]
outlierindex2
```

1 (a): Yes, there are outliers in the dataset.    
1 (b): no, the isolation forest and LOF don't agree each other. I think the reason why they disagree is that the ways they detect outliers are different. LOF will draw circles from each point and see how many other points are included. This is called k numbers. IsoForest tries to draw random lines to sepearte data points. If a points takes fewer lines to seperate, it would be an outlier.    
1 (c): The local outlier factor (LOF) method scores points in a multivariate dataset whose rows are assumed to be generated independently from the same probability distribution.      
The isolation-based approach measures how isolated a point is without any assumption on the data distribution  
  
Since I don't know the distribution of this data or how they are generated, I will use isolation forest to detect the outliers.

##### remove outliers detected by isolation forest

```python
df_droped = df.drop(outlierindex)
df_droped = df_droped.reset_index()
df_comp = df_droped.append(df_ERL,sort=False)
cate = df_comp['site'].astype('category')
df_comp['cate'] = cate.cat.codes
df_comp['cate'].unique()
df_comp.head()
```

```python
onevsall = []
for _, row in df_comp.iterrows():
    if row['site'] == 'CYT':
        onevsall.append(1)
    else:
        onevsall.append(0)
df_comp['onevsall'] = onevsall
```

```python
# seperate training and testing dataset
from sklearn.model_selection import train_test_split
from keras.utils import np_utils
train,test = train_test_split(df_comp, shuffle=True, train_size=0.7,test_size=0.3)
train_label = np_utils.to_categorical(train['onevsall'],num_classes=2)
test_label = np_utils.to_categorical(test['onevsall'],num_classes=2)
```

```python
train_label[2]
```

```python
print(test['cate'].unique(),train['cate'].unique())
```

We can see that CYT category responds to the second node in the ouyput layer. The splitted training and testing data both have 10 classes.

```python
from keras.models import Sequential
model = Sequential()
```

```python
from keras.layers import Dense

# input layer
model.add(Dense(units=8, activation='sigmoid', input_dim=8))

# Hidden layers
model.add(Dense(units=3, activation='sigmoid'))
model.add(Dense(units=3, activation='sigmoid'))

# output layer
model.add(Dense(units=2, activation='softmax'))
model.summary()
```

```python
# Stochastic gradient descent optimizer.
sgd = keras.optimizers.SGD(lr=0.1, momentum=0.0, decay=0.0, nesterov=False)
# compile the model
model.compile(optimizer=sgd,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
test_data = test[['mcg', 'gvh', 'alm', 'mit', 'erl', 'pox', 'vac', 'nuc']]
```

```python
class WeightHistory(keras.callbacks.Callback):

    def __init__(self, model, wt, error, logs = {}):

        self.weights = wt
        self.model = model
        self.error = error


    def on_epoch_end(self, epoch, logs = {}):
        self.weights.append(self.model.layers[3].get_weights())
        self.error.append(self.model.evaluate(test_data, test_label))
```

```python
Weights = []
error = []
callback = WeightHistory(model, Weights, error)
```

```python
train_data = train[['mcg', 'gvh', 'alm', 'mit', 'erl', 'pox', 'vac', 'nuc']]
hist = model.fit(train_data, train_label, epochs=20, callbacks = [callback])
```

```python
w1, w2, w3, b1 = [], [], [], [],
for i in range(0, 19):
    w1.append(callback.weights[i][0][0][1]) # weight 1
    w2.append(callback.weights[i][0][1][1]) # weight 2
    w3.append(callback.weights[i][0][2][1]) # weight 3
    b1.append(callback.weights[i][1][1])
```

```python
#plot
plt.plot(w1)
plt.plot(w2)
plt.plot(w3)
plt.plot(b1)
plt.legend(['weight1', 'weight2', 'weight3', 'bias'])
plt.xlabel('epoch')
plt.ylabel('weights')
```

```python
test_error = [1-i[1] for i in callback.error]
training_error = [1 - i for i in  hist.history['acc']]
plt.plot(training_error)
plt.plot(test_error)
plt.xlabel('epoch')
plt.ylabel('error')
plt.legend(['training_error', 'testing_error'])
```

## 5. 


```python
def create_neuronet(num_layer, num_node):
    model = Sequential()
    # input layer
    model.add(Dense(units=8, activation='sigmoid', input_dim=8))

    # Hidden layers
    for i in range(num_layer):
        model.add(Dense(units=num_node, activation='sigmoid'))

    # output layer
    model.add(Dense(units=10, activation='softmax'))

    # Stochastic gradient descent optimizer.
    sgd = keras.optimizers.SGD(lr=0.1, momentum=0.0, decay=0.0, nesterov=False)
    # compile the model
    model.compile(optimizer=sgd,
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    model.fit(train_data, train_label, epochs=1000,verbose=0)
    result = model.evaluate(test_data,test_label)
    return 1 - result[1]
```

```python
for i in range(1,4):
    for j in range(3,13,3):
        print('the testing error for {} hidden layers and {} nodes is {}'.format(i,j,create_neuronet(i, j)))
```
